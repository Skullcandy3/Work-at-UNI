```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

# A utility function to do inorder traversal of BST
def inorder(root):
    if root is not None:
        inorder(root.left)
        print(root.key, end=' ')
        inorder(root.right)

# A utility function to insert a new node with given key in BST
def insert(node, key):
    # If the tree is empty, return a new node
    if node is None:
        return Node(key)
    # Otherwise, recur down the tree
    if key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)

    # return the (unchanged) node pointer
    return node

leaf_count = 0
def count_leaf_nodes(root):
    """ Let us take the following BST
      50
    / \
    30     70
    / \ / \
  20 40 60 80
  Then it is expected to print 4 by calling the function
    """
    # write your code here
    global leaf_count
    if root is None:
        return
    if root.left is None and root.right is None:
        leaf_count += 1
    count_leaf_nodes(root.left)
    count_leaf_nodes(root.right)
    # end

def preorder(root):
    """ Let us take following BST
      50
    / \
    30     70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,20,40,70,60,80 by calling the function
    """
    # write your code here
    if root is not None:
        print(root.key, end=',')
        preorder(root.left)
        preorder(root.right)
    # end

def print_nodes_with_two_child(root):
    """ Let us take following BST
      50
    / \
    30     70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,70 by calling the function
    """
    # write your code here
    if root is None:
        return
    if root.left is not None and root.right is not None:
        print(root.key, end=',')
    print_nodes_with_two_child(root.left)
    print_nodes_with_two_child(root.right)
    # end

def print_nodes_bylevel(root):
    """ Let us take following BST
      50
    / \
    30     70
    / \ / \
  20 40 60 80

  Then it is expected to print 50,30,70,20,40,60,80 by calling the function
    """
    # write your code here
    if root is None:
        return
    queue = [root]
    while queue:
        current = queue.pop(0)
        print(current.key, end=',')
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    # end

def tree_height(root):
    # write your code here
    if root is None:
        return 0
    left_height = tree_height(root.left)
    right_height = tree_height(root.right)
    return max(left_height, right_height) + 1
    # end

# Driver Code
if __name__ == '__main__':

    #Q1:
    root = None

    # insert all nodes in the given order    by calling insert function
    nodes = [50,30,20,40,70,60,80,71,25,85]
    for key in nodes:
        root = insert(root, key)

    #Q2:
    print("\n tree_height:")
    print(tree_height(root))

    #Q3:
    print("\n Number of leaf nodes:")
    leaf_count = 0
    count_leaf_nodes(root)
    print(leaf_count)

    #Q4:
    print("\n print nodes in level order:")
    print_nodes_bylevel(root)

    #Q5:
    print("\n preorder traversal of tree:")
    preorder(root)

    #Q6:
    print("\n print_nodes_with_two_child:")
    print_nodes_with_two_child(root)
```
