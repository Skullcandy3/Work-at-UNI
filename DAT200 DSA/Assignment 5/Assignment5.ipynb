**Part 1**: Your objective in this assignment is to implement a “My_HashTable class” that supports the following operations:

•	Insert: Insert a new key-value pair

•	Search: Find the value associated with a key

•	Update: Update the value associated with a key

•	display_all: List all the keys stored in the hash table

Complete the hash table implementation below by following the instructions in the comments.

 (30 p)

```python
class MyHashTable:
    def __init__(self):
        self.max_size = 4096
        self.data = [None] * self.max_size

    def get_hash(self, key):
        hash_value = sum(ord(char) for char in key)
        return hash_value

    def get_index(self, key):
        return self.get_hash(key) % self.max_size

    def insert(self, key, value):
        index = self.get_index(key)
        self.data[index] = (key, value)

    def search(self, key):
        index = self.get_index(key)
        pair = self.data[index]
        if pair and pair[0] == key:
            return pair[1]
        return None

    def update(self, key, value):
        index = self.get_index(key)
        if self.data[index] and self.data[index][0] == key:
            self.data[index] = (key, value)
        else:
            self.insert(key, value)

    def display_all(self):
        result = []
        for item in self.data:
            if item:
                result.append(item)
        return result
    

# Reading CSV with pandas
import pandas as pd

df = pd.read_csv('Phonebook.csv', dtype={'phone': str})

Hashing = MyHashTable()
for _, row in df.iterrows():
    Hashing.insert(row['phone'], row[' name'])


# Display all entries
print(Hashing.display_all())
print(Hashing.get_hash("3251273836"))

# Adding part 3 into hashtable
phone_number_1 = '123456789'
name_1 = 'UiS'

phone_number_2 = '987654321'
name_2 = 'NTNU'

Hashing.insert(phone_number_1, name_1)
print(Hashing.display_all())
Hashing.insert(phone_number_2, name_2)
print(Hashing.display_all())
# OH NO A COLLISION BETWEEN UiS and NTNU 
# This is because they have the same hash value so UiS is overwritten by NTNU
```

Handling Collisions with Linear Probing: As you might have wondered, multiple keys can have the same hash.

Data stored against one key may override the data stored against another, if they have the same hash.

```python
class ProbingHashTable:
    def __init__(self):
        self.max_size = 4096
        self.data = [None] * self.max_size  # Corrected: initialize after setting max_size

    def get_hash(self, key):
        return sum(ord(char) for char in key)

    def get_index(self, key):
        return self.get_hash(key) % self.max_size

    def get_valid_index(self, key):
        index = self.get_index(key)
        start_index = index

        while self.data[index] is not None and self.data[index][0] != key:
            index = (index + 1) % self.max_size
            if index == start_index:
                raise Exception("Hash table is full")
        return index

    def insert(self, key, value):
        index = self.get_valid_index(key)
        self.data[index] = (key, value)

    def search(self, key):
        index = self.get_index(key)
        start_index = index

        while self.data[index] is not None:
            if self.data[index][0] == key:
                return self.data[index][1]
            index = (index + 1) % self.max_size
            if index == start_index:
                break
        return None

    def update(self, key, value):
        index = self.get_valid_index(key)
        self.data[index] = (key, value)

    def display_all(self):
        return [item for item in self.data if item is not None]
    
# Reading CSV with pandas
import pandas as pd

df = pd.read_csv('Phonebook.csv', dtype={'phone': str})

ProbHashing = ProbingHashTable()
for _, row in df.iterrows():
    ProbHashing.insert(row['phone'], row[' name'])

# Display all entries
print(ProbHashing.display_all())

# Adding last part into probing hashtable
phone_number_1 = '123456789'
name_1 = 'UiS'

phone_number_2 = '987654321'
name_2 = 'NTNU'

ProbHashing.insert(phone_number_1, name_1)
print(ProbHashing.display_all())
ProbHashing.insert(phone_number_2, name_2)
print(ProbHashing.display_all())
# Jippi there is no Collision and UiS is not overwritten by NTNU entry.
```
